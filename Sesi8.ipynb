```python
import pandas as pd
import numpy as np
```

```python
df_can = pd.read_excel('datasets/Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_can.rename(columns={'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'}, inplace=True)

df_can['Total'] = df_can.sum(axis=1)

df_can.set_index('Country', inplace=True)

df_can.columns = list(map(str, df_can.columns))

years = list(map(str, range(1980, 2014)))

df_can.shape
```

```python
df_can.head()
```

```python
df_can.reset_index(inplace=True)
```

```python
df_can.to_csv('Canada_cleaned.csv', index=False)
```

# WordClouds

```python
# !pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('datasets/alice_novel.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=2000
)

alice_wc.generate(alice_novel)
```

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')

alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
alice_mask = np.array(Image.open('datasets/alice_mask.png'))
```

```python
alice_wc = WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=2000,
    mask=alice_mask
)

alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
total_immigration = df_can['Total'].sum()

max_words = 10000
word_string = ''
for country in df_can.index.values:
    # check if country's name is a single-word name
    if len(country.split(' ')) == 1:
        repeat_num_times = int(df_can.loc[country, 'Total']/float(total_immigration)*max_words)
        word_string = word_string + ((country + ' ') * repeat_num_times)
```

```python
wordcloud = WordCloud(background_color='white').generate(word_string)
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Regression Plot with Seaborn

```python
# we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

# change the years to type float (useful for regression later on)
df_tot.index = map(float,df_tot.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['year', 'total']

# view the final dataframe
df_tot.head()
```

```python
ax = sns.regplot(x='year', y='total', data=df_tot)
```

```python
ax = sns.lineplot(x='year', y='total', data=df_tot)
```

```python
ax = sns.regplot(x='year', y='total', data=df_tot, color='green')
```

```python
ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+')
```

```python
plt.figure(figsize=(15, 10))
ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+')
```

```python
plt.figure(figsize=(15, 10))
ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+')

ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 to 2013')

plt.show()
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)
ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+', scatter_kws={'s': 200})

ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 to 2013')

plt.show()
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)
sns.set_style('ticks')

ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+', scatter_kws={'s': 200})

ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 to 2013')

plt.show()
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)
sns.set_style('whitegrid')

ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+', scatter_kws={'s': 200})

ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 to 2013')

plt.show()
```

# Maps With Marker

```python
df_incidents = pd.read_csv('datasets/Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df_incidents.head()
```

```python
df_incidents.shape
```

```python
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=13)

sanfran_map
```

```python
incidents = folium.map.FeatureGroup()
```

```python
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill='True',
            fill_color='blue',
            fill_opacity=0.6
        )
    )
```

```python
sanfran_map.add_child(incidents)
```

```python
incidents = folium.map.FeatureGroup()
```

```python
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill='True',
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)
```

```python
sanfran_map.add_child(incidents)
```

```python
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# loop through the 100 crimes and add each to the map
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)

# show map
sanfran_map
```

```python
from folium import plugins
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = plugins.MarkerCluster().add_to(sanfran_map)

for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
    location=[lat, lng],
    icon=None,
    popup=label
    ).add_to(incidents)
    
sanfran_map
```
